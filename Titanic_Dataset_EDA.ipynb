Step 4: Visualizing Key Insights


Distribution by Sex


```python
sns.countplot(x='Sex', data=titanic_df)
plt.title('Gender Distribution')
```

Distribution by Age

```python
sns.histplot(titanic_df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
```

Distribution by Passenger Class


```python
sns.countplot(x='Pclass', data=titanic_df)
plt.title('Passenger Class Distribution')
```

Distribution of Fare by Passengers

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(data=titanic_df, x='Fare', bins=20, kde=True)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

Survival by Age

```python
sns.boxplot(x='Survived', y='Age', data=titanic_df)
plt.title('Age vs Survival')
```

Passenger Class Distribution by Sex

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic_df, x='Pclass', hue='Sex', palette='Set2')
plt.title('Passenger Class Distribution by Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Sex')
plt.show()
```

Correlation Matrix for Numerical Features


```python
corr_features = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
corr_matrix = titanic_df[corr_features].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
